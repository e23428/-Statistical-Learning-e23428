# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.


1. In this model, the probability of the point falling into a region is directly proportional to the area of that region.

Total Area of the Disk:$S_{total} = \pi R^2 = \pi (10)^2 = 100\pi$

Area of Event A:Event $A$ is the "crust" of the disk, an annulus with an outer radius of $10$ and an inner radius of $9$.

$S_A = \pi (10)^2 - \pi (9)^2 = 100\pi - 81\pi = 19\pi$Probability $P(A)$:$$P(A) = \frac{Area(A)}{Area(Total)} = \frac{19\pi}{100\pi} = 0.19$$

2. In this model, the distance $r$ and the angle $\theta$ are chosen independently. Here, the probability depends only on the length of the interval of the radius.

Radial Distribution:Since $r$ is chosen uniformly from $[0, 10]$, the probability density function is $f(r) = \frac{1}{10}$.

Condition for Event A:The point lies within 1 unit of the boundary if $9 \le r \le 10$.

Probability $P(A)$:$$P(A) = P(9 \le r \le 10) = \int_{9}^{10} \frac{1}{10} \, dr = \left[ \frac{r}{10} \right]_{9}^{10} = \frac{10}{10} - \frac{9}{10} = 0.1$$

3. The discrepancy arises because "choosing a point at random" is ambiguous until you define the measure being used.

Area vs. Distance:The first model assumes every "patch" of space is equally likely. Because the outer rings of a disk have more area than the inner rings, the area model "weights" the boundary more heavily.

The Density Shift:In the second model, you are picking a distance $r$. Picking $r=1$ is just as likely as picking $r=9$. However, there is much more physical area available at $r=9$ than at $r=1$. By making the radius uniform, you are effectively "crowding" more points toward the center compared to a truly uniform spatial distribution.

Mathematical Relationship:In the area model, the probability density of the radius is actually $f(r) = \frac{2r}{R^2}$. This means the probability of picking a certain radius increases as you move outward. The second model ignores this geometric growth, leading to a lower probability ($0.1$ vs $0.19$).

# Q2) Bertrand's Paradox

A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

Show that different mathematically natural models of “random chord” lead to different answers.

Model 1: Random Endpoints

In this model, we choose two points independently and uniformly on the circumference of the circle and connect them.

Logic: Fix one point $A$ as a vertex of the inscribed equilateral triangle. To make the chord longer than the side of the triangle, the second point $B$ must fall on the arc between the other two vertices of the triangle.

Calculation: The triangle divides the circumference into three equal arcs of $120^\circ$ each. The "favorable" arc is the one opposite point $A$.

Result:$$P = \frac{1}{3} \approx 0.333$$

Model 2: Random Radial Distance

In this model, we choose a radius of the circle and then choose a point at random along that radius to serve as the midpoint of the chord.

Logic: A chord is uniquely determined by its midpoint. If we pick a random radius and then a random point at distance $r$ from the center, the chord length exceeds $L$ only if $r < R/2$.

Calculation: Since $r$ is chosen uniformly from the interval $[0, R]$, the probability is the ratio of the favorable length ($0$ to $R/2$) to the total length ($0$ to $R$).

Result:$$P = \frac{R/2}{R} = \frac{1}{2} = 0.5$$

Model 3: Random Midpoint (Uniform Area)

In this model, we choose a point anywhere inside the circle at random (uniformly by area) and let that point be the midpoint of the chord.

Logic: As established, a chord is longer than the triangle's side if its midpoint lies within a distance of $R/2$ from the center. This defines a smaller, concentric circle (the "inner disk").

Calculation: The probability is the ratio of the area of the inner disk (radius $R/2$) to the area of the entire circle (radius $R$).

Area of inner disk: $\pi (R/2)^2 = \frac{1}{4}\pi R^2$

Area of total disk: $\pi R^2$

Result:$$P = \frac{\frac{1}{4}\pi R^2}{\pi R^2} = \frac{1}{4} = 0.25$$

#Q8) Probability of Being a Girl?

What is the probability that both children are girls, given that **at least one** of them is a girl?

The Sample Space


There are four equally likely outcomes for two children:

BB (Boy, Boy)

BG (Boy, Girl)

GB (Girl, Boy)

GG (Girl, Girl)

Apply the Condition


We are given that at least one of them is a girl. This information changes our "world" by eliminating the possibility that both are boys.

Possible outcomes now: {BG, GB, GG}

Total number of outcomes in this reduced set: 3

Identify the Favorable Outcome


From our reduced set of three possibilities, we want to find the probability that both are girls.

Favorable outcome: {GG}

Number of favorable outcomes: 1

Result$$P(\text{Both Girls} \mid \text{At least one Girl}) = \frac{1}{3}$$

Many people intuitively guess $1/2$, thinking, "We already know one is a girl, so we're just looking at the other child, who has a 50/50 chance.

"However, that logic only applies if you were told a specific child (like the "elder" or the "one in the pink shirt") is a girl. By saying "at least one," we aren't specifying which child is the girl. We are only excluding the BB scenario, leaving us with three equally likely scenarios (BG, GB, GG), only one of which satisfies the "both girls" condition.

1. The Probability Space $(\Omega, \mathcal{F}, P)

$Sample Space:

$\Omega = \{G_1, G_2, D_1, D_2\}$

$\sigma$-algebra:

$\mathcal{F} = \mathcal{P}(\Omega)$ (the power set). This contains all $2^4 = 16$ possible subsets of $\Omega$:$\{\emptyset, \{G_1\}, \{G_2\}, \{D_1\}, \{D_2\}, \{G_1, G_2\}, \dots, \Omega\}

Probability Measure P:

$P(G_1) = 0.50$, $P(G_2) = 0.20$, $P(D_1) = 0.10$, $P(D_2) = 0.20$

 2. $\sigma$-algebras Generated by $X$ and $Y$A $\sigma$-algebra generated by a random variable consists of all sets of outcomes that the variable can distinguish.

 $\sigma(X)$: $X$ only cares if an item is Good ($X=0$) or Defective ($X=1$).Events: $\{G_1, G_2\}$ (Good) and $\{D_1, D_2\}$ (Defective).$\sigma(X) = \{\emptyset, \{G_1, G_2\}, \{D_1, D_2\}, \Omega\}$

 $\sigma(Y)$: $Y$ assigns a unique value to every single outcome in $\Omega$.Since every singleton is "visible" to $Y$, $\sigma(Y)$ is the full power set $\mathcal{P}(\Omega)$.

 $\sigma(Y) = \mathcal{F}$

 Every set in $\sigma(X)$ is also in $\sigma(Y)$, but not vice versa. $Y$ contains more information because knowing $Y$ tells you exactly which machine produced the item and its quality. Knowing $X$ only tells you the quality; the machine identity is "lost" or blurred.

 3. The random variable $Y$ records both quality and machine:

 The mapping is defined as follows:$$Y(G_1) = 1, \quad Y(G_2) = 2, \quad Y(D_1) = 3, \quad Y(D_2) = 4$$Since each outcome is assigned a distinct value, $Y$ distinguishes every point of $\Omega$. Therefore, the $\sigma$-algebra generated by $Y$ is the power set of the sample space:$$\sigma(Y) = \mathcal{P}(\Omega)$$Explicitly, this contains all $2^4 = 16$ possible subsets of $\Omega$:$$\sigma(Y) = \{ \emptyset, \{G_1\}, \{G_2\}, \{D_1\}, \{D_2\}, \{G_1, G_2\}, \{G_1, D_1\}, \dots, \Omega \}$$

 4. Conditional Probabilities

 $P(X=1 \mid Y=3)$: If we know $Y=3$ (Machine 1, Defective), the probability that it is defective ($X=1$) is 1.0.

 $P(X=1 \mid Y=4)$: Similarly, if $Y=4$, then $X$ must be 1. Probability = 1.0.

 $P(Y=3 \mid X=1)$: Given it is defective, what is the chance it came from Machine 1?$$\frac{P(Y=3, X=1)}{P(X=1)} = \frac{0.1}{0.3} = \mathbf{1/3}$$$P(Y=4 \mid X=1)$: Given it is defective, what is the chance it came from Machine 2?$$\frac{0.2}{0.3} = \mathbf{2/3}$$

 Interpretation: Conditioning on $Y$ provides "certainty" about $X$ because $Y$ is a finer measurement. Conditioning on $X$ still leaves uncertainty about the specific machine (the sub-categories of $Y$).

 5. Information Analysis

 What is in $\sigma(Y)$ but not $\sigma(X)$? Information about which machine produced the item. $\sigma(X)$ cannot distinguish between $G_1$ and $G_2$.

 What is in $\sigma(X)$ but not $\sigma(Y)$? Nothing. Since $\sigma(X) \subseteq \sigma(Y)$, everything $X$ knows, $Y$ also knows. In information theory terms, $X$ is a "coarsening" of $Y$.

 6. The marginal distribution of $Y$ isThe random variable $Y$ maps each outcome in $\Omega$ to a unique integer. Therefore, the probability of each value of $Y$ is simply the probability of the corresponding single outcome:$$P(Y = 1) = P(\{G_1\}) = 0.50,$$$$P(Y = 2) = P(\{G_2\}) = 0.20,$$$$P(Y = 3) = P(\{D_1\}) = 0.10,$$$$P(Y = 4) = P(\{D_2\}) = 0.20.$$

 7. Shannon Entropy (bits)

 Using the formula $H(V) = -\sum p_i \log_2(p_i)
 $:

 Joint $H(X, Y)$:Since $Y$ determines $X$, the joint entropy is just the entropy of the most descriptive variable:$H(X, Y) = H(Y) \approx \mathbf{1.761 \text{ bits}}$

 Conditional $H(Y|X)$:The uncertainty remaining in $Y$ after we know $X$:$H(Y|X) = H(X, Y) - H(X) = 1.761 - 0.881 = \mathbf{0.880 \text{ bits}}$

 Conditional $H(X|Y)$:Since $Y$ tells us everything about $X$, there is zero uncertainty left:$H(X|Y) = \mathbf{0 \text{ bits}}$

 8.  Compute the Conditional Probabilities

 To find these, we use the formula $P(A|B) = \frac{P(A \cap B)}{P(B)}$.

 $P(X=1 \mid Y=3)$:If we know $Y=3$, we know the outcome is $\{D_1\}$.  9. Interpret the Difference in Conditioning

 The difference highlights the direction of information flow:

 Conditioning on $Y$ provides "perfect" information about $X$. Because $Y$ is more granular (it tracks both machine and quality), knowing the value of $Y$ completely collapses any uncertainty about $X$.

 Conditioning on $X$ only provides "partial" information about $Y$. Knowing an item is defective ($X=1$) narrows down the possibilities for $Y$, but it doesn't tell you which machine produced it. You are left with a probabilistic distribution across the possible machines.

 10. Information in $\sigma(X)$ vs $\sigma(Y)

 Available in $\sigma(Y)$ but not $\sigma(X)$:The specific Machine Identity. In $\sigma(Y)$, the events $\{G_1\}$ and $\{G_2\}$ are distinct. In $\sigma(X)$, they are "blurred" into a single set $\{G_1, G_2\}$. $\sigma(Y)$ allows you to ask "Was this made by Machine 1?", whereas $\sigma(X)$ does not have the resolution to answer that.

 Available in $\sigma(X)$ but not $\sigma(Y)$:Nothing. Since $\sigma(X) \subset \sigma(Y)$, every piece of information available to $X$ is already contained within $Y$. In mathematical terms, $X$ is a measurable function with respect to $\sigma(Y)$. If you have the data for $Y$, you can always derive $X$.

1. The Probability Space $(\Omega, \mathfrak{F}, P)$

The underlying probability space is defined as:

Sample Space: $\Omega = \mathbb{R}^2$, where an element $\omega = (t, p)$ represents standardized temperature and pressure deviations.

$\sigma$-algebra: $\mathfrak{F} = \mathfrak{B}(\mathbb{R}^2)$, which is the Borel $\sigma$-algebra on the plane.

Probability Measure: $P$ is induced by the joint density function:$$f_{T,P}(t,p) = \frac{1}{4\pi}\exp\left( -\frac{t^2}{2} - \frac{p^2}{8} \right), \quad (t,p) \in \mathbb{R}^2$$For any Borel set $A \subseteq \mathbb{R}^2$, $P(A) = \iint_{A} f_{T,P}(t,p) dt dp$.

2. $\sigma$-algebra generated by $X$

The measurement $X$ is defined by $X(t,p) = t$, recording only temperature.Since $X^{-1}(B) = \{(t,p) \in \mathbb{R}^2 : t \in B\} = B \times \mathbb{R}$, we obtain:$$\sigma(X) = \{ B \times \mathbb{R} : B \in \mathfrak{B}(\mathbb{R}) \}$$This $\sigma$-algebra consists of "vertical strips" that depend solely on the temperature coordinate.

3. $\sigma$-algebra generated by $Y$

$Y$ is the identity map $Y(t,p) = (t,p)$, recording the full system state. Its generated $\sigma$-algebra is:$$\sigma(Y) = \mathfrak{B}(\mathbb{R}^2)$$Because $Y^{-1}(A) = A$ for every Borel set, $\sigma(Y)$ can distinguish any Borel subset of the plane.

4. Information Inclusion $\sigma(X) \subseteq \sigma(Y)$

Every set of the form $B \times \mathbb{R}$ is a Borel subset of $\mathbb{R}^2$, thus $\sigma(X) \subseteq \sigma(Y)$.The inclusion is strict because sets such as $[-1, 1] \times [-2, 2]$ belong to $\sigma(Y)$ but not to $\sigma(X)$ as they depend on both coordinates. This proves $Y$ contains strictly more information than $X$.

5. Marginal Density of $X$

By integrating out the pressure component $p$:$$f_X(x) = \int_{-\infty}^{\infty} \frac{1}{4\pi}\exp\left( -\frac{x^2}{2} - \frac{p^2}{8} \right) dp = \frac{1}{\sqrt{2\pi}} \exp\left( -\frac{x^2}{2} \right)$$Thus, $X \sim N(0, 1)$.

6.  Marginal Density of $Y$

Since $Y = (T, P)$, its marginal density is the joint density:$$f_Y(t,p) = \frac{1}{4\pi}\exp\left( -\frac{t^2}{2} - \frac{p^2}{8} \right)$$$Y$ is a bivariate normal random vector with independent components.

7. Computed Probabilities

$P(X \leq 0)$: By symmetry of the standard normal, $P(T \leq 0) = \mathbf{1/2}$.$P(X > 1)$: $1 - \Phi(1) \approx \mathbf{0.1587}$.$P(Y \in (-\infty, 0] \times \mathbb{R})$: Equivalent to $P(T \leq 0) = \mathbf{1/2}$.$P(Y \in [-1, 1] \times [-2, 2])$: Since $T, P$ are independent:$$P(-1 \leq T \leq 1)P(-2 \leq P \leq 2) = (2\Phi(1) - 1)^2 \approx \mathbf{0.466}$$

8. Compute the conditional distribution of $X$ given $Y = (t, p)$

Since $Y = (T, P)$ represents the full measurement of both temperature and pressure, it completely determines the value of $X$ ($T$). Therefore, the conditional distribution of $X$ given a specific observation $Y = (t, p)$ is degenerate at the value $t$:$$P(X = t \mid Y = (t,p)) = 1.$$Equivalently, the conditional law of $X$ given $Y = (t,p)$ is expressed as:$$\delta_t,$$which is the Dirac mass at $t$. This means that once both sensors are read, the temperature deviation is known with absolute certainty.

9. Compute the conditional distribution of $Y$ given $X = t$

The conditional distribution of $Y$ given $X = t$ represents the behavior of the random vector $(t, P)$ under the condition that temperature is fixed. Since $T$ and $P$ are independent, the pressure $P$ remains normally distributed with its original variance ($P \sim N(0,4)$). Thus:$$Y \mid X = t \sim (t, P).$$Formally, the conditional probability measure $P_{Y|X=t}$ is the product measure:$$\delta_t \otimes \mathcal{N}(0, 4).$$The conditional density $f_{Y|X}((u, v) \mid t)$ can be expressed as:$$f_{Y|X}((u, v) \mid t) = \delta_t(u) \cdot \frac{1}{\sqrt{8\pi}}\exp\left( -\frac{v^2}{8} \right).$$

10. Physical Interpretation of Conditioning

Conditioning on $X$: We only know the temperature; pressure remains random and uncertain.

Conditioning on $Y$: We know both temperature and pressure; all coordinates are known exactly, removing all uncertainty.

11. Measurability and Projections

Define the projection $\pi_1(t,p) = t$. Then $X = \pi_1 \circ Y$.Since $\pi_1$ is Borel measurable and $X$ is a measurable function of $Y$, it follows that $\sigma(X) \subseteq \sigma(Y)$, providing a formal explanation for why $Y$ is more informative.

13. Compute the differential entropy of $X$

The random variable $X$ represents the temperature deviation and follows a standard normal distribution, $X \sim N(0,1)$. The differential entropy for a normal distribution $N(\mu, \sigma^2)$ is given by the formula $h(X) = \frac{1}{2}\ln(2\pi e \sigma^2)$.Substituting $\sigma^2 = 1$:$$h(X) = \frac{1}{2}\ln(2\pi e)$$Numerical Value:$$h(X) \approx 1.4189 \text{ nats}$$

14. Compute the differential entropy of $Y$

The random variable $Y = (T, P)$ represents the full sensor state, where $T$ and $P$ are independent normal components with variances $\sigma_T^2 = 1$ and $\sigma_P^2 = 4$. For independent variables, the joint differential entropy is the sum of their individual entropies:$$h(Y) = h(T) + h(P)$$Using the entropy formula for both:$h(T) = \frac{1}{2}\ln(2\pi e)$$h(P) = \frac{1}{2}\ln(2\pi e \cdot 4) = \frac{1}{2}\ln(8\pi e)$Calculation:$$h(Y) = \frac{1}{2}\ln(2\pi e) + \frac{1}{2}\ln(8\pi e) = \ln(2\pi e) + \ln 2$$Numerical Value:$$h(Y) \approx 3.5310 \text{ nats}$$

15. Compute the conditional differential entropy $h(Y|X)$

The conditional differential entropy $h(Y|X)$ measures the uncertainty remaining in the full sensor state $Y$ once the temperature $X$ is known. Since $T$ and $P$ are independent, knowing $T$ provides no information about $P$. Therefore, the remaining uncertainty in $Y$ is exactly the uncertainty in $P$:$$h(Y|X) = h(P)$$Calculation:$$h(Y|X) = \frac{1}{2}\ln(2\pi e \cdot 4) = \frac{1}{2}\ln(8\pi e)$$Numerical Value:$$h(Y|X) \approx 2.1121 \text{ nats}$$Physical Interpretation:Once the temperature deviation is measured and known, the only remaining uncertainty in the system is the pressure deviation.

16. The $h(X \mid Y)$ Discontinuity

While in discrete cases $H(X \mid Y) = 0$ if $Y$ determines $X$, in the continuous setting, $h(X \mid Y)$ formally tends to $-\infty$. This reflects that exact deterministic dependence in continuous space corresponds to a singular distribution, highlighting a fundamental difference between discrete and differential entropy.
