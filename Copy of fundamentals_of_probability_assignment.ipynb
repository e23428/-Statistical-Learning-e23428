#Q6) Expected Project Completion Time Using PERT

A project manager is estimating the time required to complete a critical software integration task. Let $X$ denote the number of days required to complete the task. Based on expert judgment, the shortest possible time is $a=8$ days, the longest possible time is $b=20$ days, and the most likely completion time is $c=11$ days. Assume that $X$ follows a PERT distribution with parameters $(a,b,c)$.

The PERT distribution is defined as a Beta distribution scaled to the interval $[a,b]$, with internal shape parameters

$$
\alpha = 1 + 4\left( \frac{c-a}{b-a} \right)
$$

and

$$
\beta = 1 + 4\left( \frac{b-c}{b-a} \right).
$$

Its probability density function is

$$
f(x)=\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}, \qquad a \leq x \leq b,
$$

where

$$
B(\alpha,\beta)=\int_0^1 t^{\alpha-1}(1-t)^{\beta-1},dt.
$$

Answer the following:

1. Compute the values of the shape parameters $\alpha$ and $\beta$.

2. Specify the support of $X$.

3. Write down the probability density function $f_X(x)$ for this model.

4. State the probability space $(\Omega,\mathfrak F,P)$ suitable for this continuous model.

5. Define $X$ as a random variable on this space.

6. Show that the expectation $E[X]$ is given by

$$
E[X]=\frac{a+4c+b}{6}.
$$

7. Compute the variance $\mathrm{Var}(X)$ using the approximation

$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}.
$$

8. Compute the approximate standard deviation $\sigma$.

9. Briefly explain why a PERT distribution may be more appropriate than a Triangular distribution in this setting byt comparing the two results.

10. Interpret the practical meaning of the values $a$, $b$, $c$, $E[X]$, and $\sigma$ in the context of project planning.


Given values:

Optimistic time ($a$): $8$ days

Pessimistic time ($b$): $20$ days

Most likely time ($c$): $11$ days

1. Compute shape parameters $\alpha$ and $\beta$Using the formulas provided in the image:$$\alpha = 1 + 4 \left( \frac{c - a}{b - a} \right) = 1 + 4 \left( \frac{11 - 8}{20 - 8} \right) = 1 + 4 \left( \frac{3}{12} \right) = 1 + 1 = \mathbf{2}$$$$\beta = 1 + 4 \left( \frac{b - c}{b - a} \right) = 1 + 4 \left( \frac{20 - 11}{20 - 8} \right) = 1 + 4 \left( \frac{9}{12} \right) = 1 + 3 = \mathbf{4}$$

2. Support of $X$The support of a distribution is the set of all possible values the random variable can take.

Support: $[a, b] = \mathbf{[8, 20]}$

3. Probability Density Function $f_X(x)$Substitute $\alpha=2, \beta=4, a=8, b=20$ into the PDF:$$f_X(x) = \frac{(x - 8)^{2-1}(20 - x)^{4-1}}{B(2, 4)(20 - 8)^{2+4-1}} = \mathbf{\frac{(x - 8)(20 - x)^3}{B(2, 4) \cdot 12^5}}, \quad 8 \le x \le 20$$

4. Probability Space $(\Omega, \mathcal{F}, P)$

Sample Space ($\Omega$): The interval $[8, 20]$.

$\sigma$-algebra ($\mathcal{F}$): The Borel $\sigma$-algebra on $[8, 20]$, denoted as $\mathcal{B}([8, 20])$.

Probability Measure ($P$): Defined by $P(A) = \int_A f_X(x) dx$ for any $A \in \mathcal{F}$.

5. Define $X$ as a random variable

$X: \Omega \to \mathbb{R}$ is an identity mapping $X(\omega) = \omega$ for $\omega \in [8, 20]$, representing the number of days required to complete the task.

6.  Show that $E[X] = \frac{a + 4c + b}{6}$

For a Beta distribution $Y$ on $[0, 1]$ with parameters $\alpha$ and $\beta$, $E[Y] = \frac{\alpha}{\alpha + \beta}$.Since $X$ is a scaled Beta distribution: $X = a + (b - a)Y$.$$E[X] = a + (b - a) \left( \frac{\alpha}{\alpha + \beta} \right)$$Substitute $\alpha = 1 + 4\frac{c-a}{b-a}$ and $\beta = 1 + 4\frac{b-c}{b-a}$. Note that $\alpha + \beta = 2 + 4(\frac{c-a+b-c}{b-a}) = 2+4(1) = 6$.$$E[X] = a + (b - a) \left( \frac{1 + 4\frac{c-a}{b-a}}{6} \right) = \frac{6a + (b - a) + 4(c - a)}{6} = \frac{6a + b - a + 4c - 4a}{6} = \mathbf{\frac{a + 4c + b}{6}}$$Calculation: $E[X] = \frac{8 + 4(11) + 20}{6} = \frac{72}{6} = \mathbf{12 \text{ days}}$.

7. Compute Variance $\text{Var}(X)$

Using the provided approximation:$$\text{Var}(X) = \frac{(E[X] - a)(b - E[X])}{7} = \frac{(12 - 8)(20 - 12)}{7} = \frac{4 \cdot 8}{7} = \frac{32}{7} \approx \mathbf{4.57 \text{ days}^2}$$

8. Approximate Standard Deviation $\sigma$

$$\sigma = \sqrt{\text{Var}(X)} = \sqrt{\frac{32}{7}} \approx \mathbf{2.14 \text{ days}}$$

9. PERT vs. Triangular Distribution

The PERT distribution is generally more appropriate because it is smoother and places more weight on the "most likely" value ($c$) compared to the Triangular distribution. While a Triangular distribution has a "peak" that creates a sharp change in probability, PERT uses a bell-like curve (Beta distribution) that better reflects the natural uncertainty and lack of sharp "edges" in real-world human estimation.

10. Practical Interpretations

$a, b, c$: These represent the boundaries of risk (best case, worst case, and most common outcome).

$E[X]$ ($12$ days): This is the "weighted average" completion time. It’s slightly higher than the "most likely" time ($11$) because the "pessimistic" estimate ($20$) is much further from the mode than the "optimistic" one ($8$), pulling the average to the right.

$\sigma$ ($2.14$ days): Represents the risk or uncertainty level. A larger $\sigma$ indicates less confidence in the estimates. In planning, you might add $k\sigma$ to $E[X]$ to determine a "safe" deadline.
